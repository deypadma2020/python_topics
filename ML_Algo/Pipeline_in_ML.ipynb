## Example 1

### displaing simple scaling with Sandard scaling and Logistic Regression, classifier

```python
from sklearn.pipeline import Pipeline

# feature scaling
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
steps = [("standard_scaler", StandardScaler()),
         ("classifier", LogisticRegression())]

steps
```

```python
pipe = Pipeline(steps)

pipe
```

```python
# visualize Pipeline
from sklearn import set_config
set_config(display = 'diagram')

pipe
```

```python
# creating a dataset
from sklearn.datasets import make_classification
X, y = make_classification(n_samples = 1000)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.33, random_state = 42)
```

```python
X_train
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
y_pred
```

## Example 3

### Complex examples of columns transformer

#### Numerical processing Pipeline

```python
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
numeric_processor = Pipeline(
    steps = [("imputation_mean", SimpleImputer(missing_values = np.nan, strategy = "mean")),
             ("scaler", StandardScaler())]
)

numeric_processor
```

#### Categorical processing Pipeline

```python
from sklearn.preprocessing import OneHotEncoder

categorical_processor = Pipeline(
    steps = [("imputation_constant", SimpleImputer(fill_value = "missing", strategy = "constant")),
             ("onehot", OneHotEncoder(handle_unknown = "ignore"))]
)
categorical_processor
```

#### Combine processing techniques

```python
from sklearn.compose import ColumnTransformer

preprocessor = ColumnTransformer(
    [("categorical", categorical_processor,["gender", "City"]),
     ("numerical", numeric_processor, ["age", "height"])]
)
preprocessor
```

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe = make_pipeline(preprocessor, LogisticRegression())

pipe
```

## Example 4

### Hyperparameter Tuning In Pipelining Machine Learning

```python
import numpy as np

from sklearn.pipeline import make_pipeline
from sklearn.pipeline import Pipeline

from sklearn.impute import SimpleImputer

from sklearn.compose import ColumnTransformer

from sklearn.preprocessing import OneHotEncoder, StandardScaler

from sklearn.ensemble import RandomForestRegressor

from sklearn.model_selection import GridSearchCV
```

```python
import seaborn as sns

df = sns.load_dataset("tips")
df.head()
```

```python
X = df.iloc[:, 1:]
y = df['total_bill']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
numeric_preprocessor = Pipeline(
    steps =[
        ("imputation_mean", SimpleImputer(missing_values = np.nan, strategy = "mean")),
        ("scaler", StandardScaler())
    ]
)
numeric_processor
```

```python
categorical_preprocessor = Pipeline(
    steps = [
      ("imputation_constant", SimpleImputer(fill_value = "missing", strategy = "constant")),
      ("onehot", OneHotEncoder(handle_unknown = "ignore"))
    ]
)
categorical_preprocessor
```

```python
preprocessor = ColumnTransformer(
    [("categorical", categorical_preprocessor, ['sex', 'smoker', 'day', 'time']),
     ("numerical", numeric_preprocessor, ['tip', 'size'])]
)
preprocessor
```

```python
pipe = Pipeline(
    steps = [("preprocessor", preprocessor), ("regressor", RandomForestRegressor())]
)
pipe
```

```python
from sklearn import set_config
set_config(display = 'diagram')

pipe
```

```python
pipe.fit(X_train, y_train)
```

```python
pipe.predict(X_test)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

##### Hyperparameter Tuning

```python
param_grid = {
    "regressor__n_estimators" : [200, 500],
    "regressor__max_features" : ['auto', 'sqrt', 'log2'],
    "regressor__max_depth" : [4, 5, 6, 7, 8]
}
```

```python
grid_search = GridSearchCV(pipe, param_grid = param_grid, n_jobs = 1)
grid_search
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.predict(X_test)
```
